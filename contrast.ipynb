```python
import cv2
import os
import numpy as np
```

```python

image_path='bktimage'
def get_images(image_path):
    list_of_images = []
    for image_name in os.listdir(image_path):  # Loop variable renamed to 'image_name'
        if image_name.endswith('.jpg'):
            full_path = os.path.join(image_path, image_name)  # Correct usage here
            img = cv2.imread(full_path)
            list_of_images.append(img)
    return list_of_images

# Usage example
images = get_images(image_path=image_path)  # Replace with your folder path
```

```python
def get_images(image_path):
    return [cv2.imread(os.path.join (image_path,bktimage))for bktimage in os.listdir(image_path) if bktimage.endswith('.jpg')]
images=get_images(image_path=image_path)  
```

```python
hdr= cv2.createMergeMertens()
fusion_image=hdr.process(images)
```

```python
# Resize the fusion image to fit the screen
screen_res = (800 , 700)  # Replace with your screen resolution
fusion_image_resized = cv2.resize(fusion_image, screen_res)
```

```python
cv2.imshow('fusion Image',fusion_image_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
